# Irisデータセットを使った機械学習の例
## データ読み込み
データはscikitlearnのデータセットを用いる
iris.data が説明変数のデータ
iris.target が分類（教師データ）となる

```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data)
```

```python
print(iris.target)
```

```python
print(iris.data.shape)
```

```python
print(iris.target.shape)
```

## データの読み込み（Pandas）あやめの観測データ
### データの意味(iris.data)
sepal length ：　がく片の長さ

sepal width: がく片の幅

petal length: 花弁の長さ

petal width: 花弁の幅

### データの意味（iris.target)
setosa: 0

versicolor:1

virginica:2

```python
import pandas as pd
pd.DataFrame(iris.data,columns=iris.feature_names)
```

### 各項目間に関係性がないかを確認する

```python
from matplotlib import pyplot as plt
%matplotlib inline



for i, (x,y) in enumerate([(0,1),(0,2),(0,3),(1,2),(1,3),(2,3)]):
    plt.figure(figsize=(20,8))
    plt.subplot(2,3,i + 1)

    for t, marker , c in [(0,'>','r'),(1,'o','g'),(2,'x','b')]:
        plt.scatter(
            iris.data[iris.target == t,x],
            iris.data[iris.target==t,y],
            marker = marker,
            c = c,
            label = iris.target_names[t],
        )
        plt.xlabel(iris.feature_names[x])
        plt.ylabel(iris.feature_names[y])
        plt.grid()
        plt.autoscale()
        plt.legend(bbox_to_anchor=(1.05,1),loc='upper left',borderaxespad=0)
    plt.show()
        
```

## データの分離
訓練データと検証データを分離する

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(iris.data, iris.target, random_state=0)
```

## SVMによる学習

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(data_train, target_train)
```

## 学習結果の評価

```python
print('Train score: {:.3f}'.format(svc.score(data_train, target_train)))
print('Test score: {:.3f}'.format(svc.score(data_test, target_test)))
```

## 予測値の表示

```python
predicted = svc.predict(data_test)
print(predicted)
print(target_test)
```
